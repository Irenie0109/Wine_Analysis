```python
wine_main = pd.read_csv("winemag-data-130k-v2.csv")

print(wine_main.shape)

wine_main.head()
```

```python
missing_by_columns = wine_main.isnull().mean().sort_values(ascending=False).round(4)*100
missing_by_columns.plot(kind='bar',color='coral')
plt.title("% of missing data")
missing_by_columns
```

Among the columns with the most missing values are region_2(%61), designation (%29), taster_twitter_handle (%24) and taster_name (%20). So we will drop this column. The missing values in region_1 can simply mean that there is no region which can give important information

```python
#drop region_2, designation, taster_twitter_handle, taster_name
wine_column_reduced = wine_main.drop(columns=["region_2",'designation','taster_twitter_handle','taster_name'])
wine_column_reduced
```

```python
#missing by rows
#drop rows with missing price
wine_row_reduced= wine_column_reduced.dropna(subset=['country','variety','price'],how='any')
print("Number of rows, columns: ",(wine_row_reduced.shape[0],wine_row_reduced.shape[1]))
wine_row_reduced.isnull().sum().sort_values(ascending=False)
```

We will group wines by title: take mean value for the price and points, take the first entry for country, province, region_1, variety and winery (because all this attributes supposed to be the same for given title), and we will joint descriptions together making a list of avaible reviews for given wine

```python
#groups wines by title
wine_groupped=wine_row_reduced.groupby('title').agg(dict(country='first',description=list, 
                                                         points='mean',price='mean',province='first',region_1='first',
                                                         variety='first',winery='first')).reset_index()
wine_groupped
```

# 1 What countries are the major producers of wine and what sorts of wine have the best rating in those countries

```python
countries_wine_counts=wine_groupped['country'].value_counts(ascending=False)
print("Total number of countries: ",len(countries_wine_counts))
countries_wine_counts
```

We will create barplot of major wine producers that produce more than 1000 different wines

```python
countries_wine_counts[:12].plot(kind='bar',color='salmon')
plt.title('Major wine producers')
plt.ylabel('Number of wines')
plt.xlabel('Country')
major_countries=countries_wine_counts[:12].index
print("major wine producers: ",major_countries)
```

```python
wine_groupped['counts']=0
country_variety_counts=wine_groupped.groupby(['country','variety']).agg(dict(counts='count',points='median')).sort_values(by='counts',ascending=False).reset_index()
country_variety_counts[:10]
```

First we will analyse US wines and then apply similar ananlysis to other first 12 popular countries. We will choose 15 most popaular sorts (variety) and obtain mean points for it

```python
US= country_variety_counts.loc[country_variety_counts['country']=='US'][:10]
variety_list=US['variety'].values
variety_list
```

```python
wine_groupped[wine_groupped['variety'].isin(variety_list)][wine_groupped['country']=='US'].reset_index()
```

```python
#create dataframe only with US wines
df=wine_groupped[wine_groupped['variety'].isin(variety_list)][wine_groupped['country']=="US"]

#create a bosplot with median number of points indicated
fig,ax=plt.subplots(figsize=(12,6))
chart=sns.boxplot(x='variety',y='points',data=df,ax=ax,order=variety_list,color='coral')

medians=US['points'].values
median_labels=[str(int(s)) for s in medians]

pos=range(len(medians))
for tick, label in zip(pos,ax.get_xticklabels()):
    ax.text(pos[tick],medians[tick]+1.0, median_labels[tick],
           horizontalalignment='center',size='medium',color='white',weight='bold')
    plt.xticks(rotation=90)
    plt.title("Quality rating of major US varieties(left to right)")
    plt.show()
```

```python
#Create function that will determine 10 top wine varieties for a given country 
#and plot boxplot diagram of median points for these 10 varieties

def top_ten_varieties(df, country):  
    ''' Returns top ten most popular varieties by country'''
    
    country_data = df.loc[country_variety_counts['country']==country][:10]
    return country_data['variety'].values

def create_points_boxplot(df, country, n_varieties=10):
    
    '''Creates points boxplot 
    
    Input
    df: data frame
    country (str): country for which boxplot will be created
    n_varieties: number of varieties to plot
    
    
    Return: none'''
    country_data = country_variety_counts.loc[country_variety_counts['country']==country][:n_varieties]
    variety_list = country_data['variety'].values
    df = df[df['variety'].isin(variety_list)][df['country']==country]

    fig, ax = plt.subplots(figsize = (10,9))
    chart = sns.boxplot(x='variety',y='points', data=df, ax = ax, order=variety_list,color='salmon')

    medians = country_data['points'].values
    median_labels = [str(int(s)) for s in medians]

    pos = range(len(medians))
    for tick,label in zip(pos,ax.get_xticklabels()):
        ax.text(pos[tick], medians[tick] + 0.3, median_labels[tick], 
                horizontalalignment='center', size='medium', color='white', weight='semibold')
    plt.xticks(rotation = 90)
    plt.title(country)
    plt.show()
```

```python
#Create boxplot for 12 major countries with 12 most popular varieties
# x axis from left to right from most popular to least popular
import warnings
warnings.filterwarnings('ignore')
for country in major_countries:
    create_points_boxplot(wine_groupped,country,15)
```
